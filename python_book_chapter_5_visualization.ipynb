```python
#  کتابخانه های زیر را بارگذاری می کنیم

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
from pylab import *
```

```python
# تابع زیر را رسم خواهیم کرد
x = np.linspace(-5, 5, 200)
y1 = x**2
y2 = x**3
```

```python
#پلات نمایش تابع 

fig, ax = plt.subplots()
ax.plot(x, y1, 'r' ,label=r"$y_1 = x^2$", linewidth=2)
ax.plot(x, y2, 'k--',label=r"$y_2 = x^3$", linewidth=2)
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$x$', fontsize=18)
ax.set_ylabel(r'$y$', fontsize=18)
ax.set_title('My Figure')
plt.show()
```

```python
#دستور ذخیره کردن تصویر پلات 

fig.savefig('firstplot.png')
```

```python
#نمودار خطی 

plt.figure(figsize=(10,5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x,2)
y2 = np.power(x,3)
plt.plot(x,y1, "b-" , label = '$y1 = x^2$') # Setting up legends
plt.plot(x, y2,color='red',linewidth=1.0,linestyle='--') # Setting up legends
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.legend(loc='upper center', fontsize='large')
plt.show()
```

```python

plt.figure(figsize=(12,6))
x = np.linspace(0, 10, 100)
y1 = np.sin(x) 
y2 = np.cos(x) 
plt.subplot(2,1,1)
plt.plot(x,y1, "b-")
plt.subplot(2,1,2)
plt.plot(x,y2, "r-")
plt.tight_layout()
plt.show()
```

```python
x = np.arange(1,21)
plt.figure(figsize=(16,8))
y1 = np.random.uniform(0.1,0.7,20)
y2 = np.random.uniform(0.1,0.7,20)

plt.bar(x, +y1, facecolor='#C0CA33', edgecolor='white') #specify edgecolor by name
plt.bar(x, -y2, facecolor='#FF9800', edgecolor='white')

for x,y in zip(x,y1):
   plt.text(x, y+0.05, '%.2f' % y, ha='center' , va= 'bottom', fontsize = 10)
   
plt.xlim(0,21)
plt.ylim(-1.25,+1.25)
plt.show()
```
